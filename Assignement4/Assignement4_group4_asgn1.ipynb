## Group
How much of kg / t / kt / kt CO2-eqv. are produces by each sector ?  
There are 4 different units kt CO2-eqv, kt, t, kg so we are gonna divide them.

There are already rows for the total by sector so we are gonna use them.

```r
options(repr.plot.width = 20)
plot_emissions <- function(df, unit) {
    filtered_df <- df %>%
        filter(str_detect(greenhouse.gas, fixed(unit, ignore_case = TRUE)))
    
    long_df <- filtered_df %>%
        pivot_longer(
            cols = starts_with("X"),
            names_to = "Year",
            values_to = "Emissions"
        ) %>%
        mutate(
            Year = str_remove(Year, "^X")
        ) %>%
        mutate(Year = as.integer(Year)) %>%
        mutate(greenhouse.gas = factor(greenhouse.gas, levels = rev(unique(greenhouse.gas))))
        
    
    ggplot(long_df, aes(x = Year, y = Emissions, fill = greenhouse.gas)) +
        geom_area(alpha = 0.6) +
        labs(title = "Emissions by Gas Over Time", x = "Year", y = "Emissions") +
        theme(
            plot.title = element_text(hjust = 0.5,size = 25,margin = margin(t = 20, b = 20) ),
            axis.text.x = element_text(size = 16),
            axis.title.x = element_text(size = 20, margin = margin(t = 20, b = 20), ),
            axis.text.y = element_text(size = 16),
            axis.title.y = element_text(size = 20, margin = margin(r = 20, l = 20), ),
            legend.title = element_text(size = 18),
            legend.title.align = 0.5,
            legend.text = element_text(size = 16),
            panel.grid = element_line(size = 1.5)
        ) +
        scale_x_continuous(breaks = scales::breaks_width(5))
}

rm_tot <- df %>%
    filter(greenhouse.gas != "Total Greenhouse Gases (kt CO2-eqv.)")

grouped <- rm_tot %>%
    group_by(greenhouse.gas) %>%
    summarise(across(starts_with("X"), ~ sum(.x, na.rm = TRUE)), .groups = "drop")


plot_emissions(grouped, "(kg)")
plot_emissions(grouped, "(t)")
plot_emissions(grouped, "(kt)")
plot_emissions(grouped, "(kt CO2-eqv.)")
```

```r
df_filtered <- df %>%
  filter(greenhouse.gas == "Total Greenhouse Gases (kt CO2-eqv.)") %>%
  select(-greenhouse.gas)

df_filtered_no_nt <- df_filtered %>%
  filter(sector != "NATIONAL TOTAL (including LULUCF, excluding international transports)")

long_df <- df_filtered_no_nt %>%
  pivot_longer(cols = starts_with("X"), names_to = "Year", values_to = "Emissions") %>%
  mutate(
    Year = str_remove(Year, "^X"),
    Year = as.integer(Year)
  ) %>%
  mutate(sector = factor(sector, levels = rev(unique(sector))))

ggplot(long_df, aes(x = Year, y = Emissions, fill = sector)) +
  geom_area(alpha = 0.6) +
  scale_fill_viridis_d() +
  labs(
    title = "Emissions by Sector Over Time",
    x = "Year",
    y = "Emissions",
    fill = "Sector"
  ) +
  theme(
    plot.title = element_text(hjust = 0.5,size = 25,margin = margin(t = 20, b = 20) ),
    axis.text.x = element_text(size = 16),
    axis.title.x = element_text(size = 20, margin = margin(t = 20, b = 20), ),
    axis.text.y = element_text(size = 16),
    axis.title.y = element_text(size = 20, margin = margin(r = 20, l = 20), ),
    legend.title = element_text(size = 18),
    legend.title.align = 0.5,
    legend.text = element_text(size = 16),
    panel.grid = element_line(size = 1.5)
  ) +
  scale_x_continuous(breaks = scales::breaks_width(5))
```

### Isolate waste data

I want to isolate waste data, work on the data to know which gases are waste. Saw the evolution between 1990 and 2023

```r
(df_waste_total <- df %>%
    filter(sector == "WASTE, TOTAL") %>%
    arrange(desc(`X2023`)))
```

#### Saw the evolution between 1990 and 2023

```r
(df_waste_total_only_1990_2023 <- df_waste_total %>% 
    select(`greenhouse.gas`, sector, `X1990`, `X2023`))
```

```r
(df_waste_total_only_1990_2023_with_evolution <- df_waste_total_only_1990_2023 %>%
    mutate(percent_change = ((`X2023` - `X1990`) / `X1990`) * 100) %>%
    replace_na(list(percent_change = 0)) %>% 
    filter(`X1990` != 0, `X2023` != 0) %>%
    arrange(desc(percent_change)))
```

#### plotting the evolution

```r
df_waste_total_only_1990_2023_with_evolution <- df_waste_total_only_1990_2023_with_evolution %>%
    mutate(`greenhouse.gas` = fct_reorder(`greenhouse.gas`, percent_change))

ggplot(df_waste_total_only_1990_2023_with_evolution, aes(x = percent_change, y = `greenhouse.gas`)) +
    geom_col(fill = "mediumseagreen") +
    labs(
    x = "Percent change (%)",
    y = "Greenhouse Gas",
    title = "Percent change of emissions from Waste from 1990 to 2023"
    ) +
    theme_minimal() +
    theme(
        plot.title = element_text(hjust = 0.5,size = 25,margin = margin(t = 20, b = 20) ),
        axis.text.x = element_text(size = 16),
        axis.title.x = element_text(size = 20, margin = margin(t = 20, b = 20), ),
        axis.text.y = element_text(size = 16),
        axis.title.y = element_text(size = 20, margin = margin(r = 20, l = 20), ),
        legend.title = element_text(size = 18),
        legend.title.align = 0.5,
        legend.text = element_text(size = 16),
        panel.grid = element_line(size = 1.5)
    ) +
    scale_x_continuous(breaks = scales::breaks_width(50))
```

## Get the top 25% of gas producers sectors by greenhouse effect

```r
df <- df %>%
  group_by(greenhouse.gas) %>%
  mutate(
    gas_quantile = quantile(Total, probs = 0.75, na.rm = TRUE),
    `Top producer` = Total >= gas_quantile
  ) %>%
  ungroup() %>%
  select(-gas_quantile)

df
```
